## Part 3

1. Describe what an identity matrix is. What does multiplying with an identity matrix do?
1. Create an identity matrix called `identity_matrix` that has the same size as `img_average_channel`.

```python
#A special type of square matrix which consists of ones(1) on it's diagonal and zeroes(0) elsewhere is known as an identity matrix.
#Any matrix remains unchanged when it is multiplied by an identity matrix of the proper size. It's similar to multiplying a number by 1

# Get the rows and columns of img_average_channel
img_rows, img_cols = img_average_channel.shape

# Create an identity matrix with the same size as img_average_channel
identity_matrix = np.eye(img_rows, img_cols)

# Print a random row from identity_matrix
random_row = identity_matrix[99,:]

print("Identity Matrix:")
print(identity_matrix)
print("\nRandom Row from Identity Matrix:")
print(random_row)
```

## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`.

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
# Create an array containing numbers 0 to N-1 where N is the number of rows in the image
rows = np.arange(img_rows)

# Shuffle the indices using numpy.random.shuffle
np.random.shuffle(rows)

# Print the shuffled array
print("Shuffled Rows:")
print(rows)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
# Use the shuffled 'rows' array to rearrange the rows of 'identity_matrix' into 'T'
T = identity_matrix[rows]

#Here, 'T' is a rearrangement of 'identity matrix'. It has the same elements as 'identity_matrix',
#but the order of the rows has been shuffled based on the values in the 'rows'

# Print the shape of 'T'
print("Shape of T:", T.shape)
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see.
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image.

```python
# Matrix multiply 'T' and 'img_average_channel' to get 'transformed_image'
transformed_image = T @ img_average_channel

# Check the shape of the result
print("Shape of transformed_image:", transformed_image.shape)

# Plot 'transformed_image'
plt.imshow(transformed_image)
plt.title('Transformed Image')
plt.show()

#Each dot product in the matrix multiplication of T and img_average_channel rearranges the rows of img_average_channel based on the
#order defined by T. It applies a permutation to the rows of the original image.By shuffling the rows, the transformation
#affects the pixels of the image. Depending on the specific shuffling
#defined by T,there will be distortions in the transformed image.
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python

# Calculate the determinant of 'T' to check if it is invertible
det_T = np.linalg.det(T)

# If 'T' is invertible, compute its inverse 'Tinv'
if det_T != 0:
    Tinv = np.linalg.inv(T)

    # Perform the inverse operation to retrieve the original image
    original_image = Tinv @ transformed_image

    # Plot the retrieved original image
    plt.imshow(original_image)
    plt.title('Original Image (Recovered)')
    plt.show()
else:
    print("Matrix 'T' is not invertible.")
```

```python
#rearange the columns and shuffle
columns=np.arange(img_cols)
np.random.shuffle(columns)

#Use your array of random column indices to create a new matrix T_col from identity_matrix
T_columns = identity_matrix[columns]

#Matrix multiply T_column and img_average_channel and store the result in transformed_image_columnar_wise
cols_transformed_image= img_average_channel @ T_columns

#Plot transformed_image_columnar_wise
plt.imshow(cols_transformed_image)
plt.title("Scrambled Image")
plt.show()
```

## Extra Points

```python
#Matrix multiply T, T_columns, and img_average_channel for rows and columns scrambled image
a= T @ img_average_channel
rows_cols_transformed_image = a @ T_columns

#Plot the scrambled image
plt.imshow(rows_cols_transformed_image)
plt.title("Scrambled Image")
plt.show()

#Compute the inverse of T_columns
T_columns_inv = np.linalg.inv(T_columns)

# Perform the inverse operation to retrieve the original image
b= Tinv @ rows_cols_transformed_image
recovered_image = b @ T_columns_inv

plt.imshow(recovered_image)
plt.title("Original Image (Recovered)")
plt.show()
```
